```python
import matplotlib.pyplot as plt

# 假設您已經有名為 "df" 的 DataFrame，且含有 "logP" 列

# 繪製散點圖
plt.scatter(df.index, df['QED'])

# 設定標籤和標題
plt.xlabel('Index')
plt.ylabel('QED')
plt.title('Scatter Plot of QED')

# 顯示圖形
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
logP_mse = mean_squared_error(df['logP'], df['ideal logP'])
tPSA_mse = mean_squared_error(df['tPSA'], df['ideal tPSA'])
QED_mse = mean_squared_error(df['QED'], df['ideal QED'])
logP_rmse = np.sqrt(logP_mse)
tPSA_rmse = np.sqrt(tPSA_mse)
QED_rmse = np.sqrt(QED_mse)

print('p_logP_rmse: %f' % logP_rmse)
print('p_tPSA_rmse: %f' % tPSA_rmse)
print('p_QED_rmse: %f' % QED_rmse)
```

```python
df['p_logP_rmse']=np.sqrt(abs(df['logP']**2-df['ideal logP']**2))
df['p_tPSA_rmse']=np.sqrt(abs(df['tPSA']**2-df['ideal tPSA']**2))
df['p_QED_rmse']=np.sqrt(abs(df['QED']**2-df['ideal QED']**2))
df['rmse_sum'] = df['p_logP_rmse']+df['p_tPSA_rmse']+df['p_QED_rmse']
```

```python
df
```

```python
df = df.sort_values(by='rmse_sum', ascending=True)
```
