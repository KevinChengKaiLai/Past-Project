```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import keras
from keras.models import Sequential
from keras.layers import Dense

```

```python
# read the data
df = pd.read_excel('Concrete-Homework.xlsx')  # 替换为你的数据文件路径
```

```python
df = df.drop('Unnamed: 0',axis = 1)
```

```python
df
```

```python
# extract X and y
features = df.drop('CompressiveStrength', axis=1)
target = df['CompressiveStrength']

# scaling the data
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)

# split train and test
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)

```

```python

# 1st NN
model_1 = Sequential()
model_1.add(Dense(10, input_dim = X_train.shape[1], activation='relu'))  # hidden layer 1 
model_1.add(Dense(1))  # output

# train
model_1.compile(loss='mean_squared_error', optimizer='adam')
model_1.fit(X_train, y_train, epochs=100, batch_size=32, verbose=1)
```

```python

# 1st NN
model_1 = Sequential()
model_1.add(Dense(10, input_dim = X_train.shape[1], activation='relu'))  # hidden layer 1 
model_1.add(Dense(1))  # output

# train
model_1.compile(loss='mean_squared_error', optimizer='adam')
model_1.fit(X_train, y_train, epochs=1000, batch_size=32, verbose=1)
```

```python
# 100 epoch
mse_1 = model_1.evaluate(X_test, y_test)

print('model 1 RMSE:', np.sqrt(mse_1))
```

```python
### 1000 epoch ###
mse_1 = model_1.evaluate(X_test, y_test)

print('model 1 RMSE with 1000 epoch:', np.sqrt(mse_1))
```

```python
# 2nd NN (2 hidden layer)
model_2 = Sequential()
model_2.add(Dense(10, input_dim=X_train.shape[1], activation='relu'))  # hidden layer1
model_2.add(Dense(10, activation='relu'))  # hidden layer 2
model_2.add(Dense(1))  # output

# train
model_2.compile(loss='mean_squared_error', optimizer='adam')
model_2.fit(X_train, y_train, epochs=100, batch_size=32, verbose=1)

```

```python
# evaluate the model

mse_2 = model_2.evaluate(X_test, y_test)

print('model 2 RMSE (1000 epoch):', np.sqrt(mse_2))
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define the paths to your training, validation, and test data directories
train_dir = 'train'
val_dir = 'valid'
test_dir = 'test'

# Define the input image size expected by the model
image_size = (224, 224)  # Adjust this based on the pretrained model's input size

# Create an instance of the ImageDataGenerator for preprocessing and data augmentation
datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

# Load and preprocess the training data
train_generator = datagen.flow_from_directory(
    train_dir,
    target_size=image_size,
    batch_size=32,  # Adjust batch size as per your preference
    class_mode='categorical')

# Load and preprocess the validation data
val_generator = datagen.flow_from_directory(
    val_dir,
    target_size=image_size,
    batch_size=32,  # Adjust batch size as per your preference
    class_mode='categorical',
    shuffle=False)

# Load and preprocess the test data
test_generator = datagen.flow_from_directory(
    test_dir,
    target_size=image_size,
    batch_size=32,  # Adjust batch size as per your preference
    class_mode='categorical',
    shuffle=False)
```

```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

# Preprocess input images to the size expected by MobileNetV2
image_size = (224, 224)  # Adjust this based on the pretrained model's input size
preprocess_input = tf.keras.applications.mobilenet_v2.preprocess_input

# Load the pretrained MobileNetV2 model without the top (fully connected) layers
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(image_size[0], image_size[1], 3))

# Add your own top layers for classification
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation='relu')(x)  # Add your desired number of neurons
x = Dense(128, activation='relu')(x) 
x = Dense(64, activation='relu')(x) 
predictions = Dense(10, activation='softmax')(x)  # Adjust the number of output classes

# Create the final model
model = Model(inputs=base_model.input, outputs=predictions)

# Freeze the base layers (optional)
for layer in base_model.layers:
    layer.trainable = False

# Compile the model
model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model on your dataset
model.fit(train_generator, epochs=5, validation_data=(val_generator ))


```

```python
test_loss, test_acc = model.evaluate(test_generator)
print('test_loss',test_loss)
print('test_acc',test_acc)
```
